```python
## essential libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing the LinearRegression class from the scikit-learn library
from sklearn.linear_model import LinearRegression

# Define the predictor variable(s) and the outcome variable

predictors = ['pm2_5'] # List of predictor variables
outcome = 'pm10' # Outcome/target variable

# Creating an instance of the LinearRegression model
model = LinearRegression()

# Fitting the model using the predictor and outcome variables
model.fit(df[predictors], df[outcome])

print(f'Intercept: {model.intercept_:.1f}') #Print the intercept rounded to 3 sf
print(f'Coefficient: {model.coef_[0]:.2f}') # Print the coefficient rounded to 3 sf
```

```python
## calculate residuals:
fitted = model.predict(df[predictors])
residuals = df[outcome] - fitted
print(residuals.head())
 
# Calculating the Root Mean Squared Error (RMSE)
from sklearn.metrics import mean_squared_error
RMSE = np.sqrt(mean_squared_error(df[outcome], fitted))
print(f'RMSE #1: {RMSE:.1f}')

# Calculating the R-squared (coefficient of determination)
from sklearn.metrics import r2_score
r2 = r2_score(df[outcome], fitted)
print(f'r2 #1: {r2:.3f}') # Print the R-squared value rounded to 3 decimal places
```

```python
# Calculate the number of observations (N) and the number of predictors (p)
N = len(df[predictors])
p = len(df[predictors].columns) + 1  # plus one because LinearRegression adds an intercept term

# Create a matrix X_with_intercept of size (N, p) filled with NaN values
# Initialize the first column with ones for the intercept term
# Initialize the remaining columns with the values of the predictor variables
X_with_intercept = np.empty(shape=(N, p), dtype=float)
X_with_intercept[:, 0] = 1
X_with_intercept[:, 1:p] = df[predictors].values
print(X_with_intercept)

# Compute the estimated coefficients (beta_hat) using the Normal Equation
beta_hat = np.linalg.inv(X_with_intercept.T @ X_with_intercept) @ X_with_intercept.T @ df[outcome].values
print(beta_hat)

# Calculate the residuals by subtracting the predicted values from the actual values
y_hat = model.predict(df[predictors])
residuals = df[outcome].values - y_hat

# Calculate the residual sum of squares (RSS)
residual_sum_of_squares = residuals.T @ residuals
print(residual_sum_of_squares)

# Estimate the variance of beta_hat and compute the standard errors for each coefficient
# Compute the estimated variance-covariance matrix of beta_hat (var_beta_hat)
# Compute the standard errors (SE) for each coefficient using the square root of the diagonal elements of var_beta_hat
sigma_squared_hat = residual_sum_of_squares / (N - p)
var_beta_hat = np.linalg.inv(X_with_intercept.T @ X_with_intercept) * sigma_squared_hat

for p_ in range(p):
    standard_error = var_beta_hat[p_, p_] ** 0.5
    print(f"SE(beta_hat[{p_}]): {standard_error}")
```

```python
import statsmodels.formula.api as sm
from statsmodels.api import add_constant

# Creating a linear regression model using the ols (ordinary least squares) method from statsmodels
# The formula specifies the relationship between the dependent variable ("pm10") and the independent variable ("pm2_5")
# The data parameter specifies the DataFrame containing the variables
mod=sm.ols(formula='df["pm10"]~df["pm2_5"]',data=df)

# Fitting the linear regression model to the data
res=mod.fit()
pd.options.display.float_format = '{:.3g}'.format
print(res.summary())
```
